```python
import tensorflow as tf
import numpy as np
import random

# Load the Fashion MNIST dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# (normalize between 0 and 1) pixel [rgb]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten input data (Input preparation layer) confert from 2d to 1d   Turn 28x28 image into 784 numbers
x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)

# Build the model with exactly 6 layers
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(784,)),               # Input layer 28*28
    tf.keras.layers.Dense(512, activation='relu'),     # Layer 2
    tf.keras.layers.Dense(256, activation='relu'),     # Layer 3
    tf.keras.layers.Dense(128, activation='relu'),     # Layer 4
    tf.keras.layers.Dense(64, activation='relu'),      # Layer 5
    tf.keras.layers.Dense(10, activation='softmax')    # Output layer
])
 #compile the model
model.compile(optimizer='adam',# optimization algo
              loss='sparse_categorical_crossentropy', # loss
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, validation_split=0.1)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"\nAdam Model Accuracy on Test Data: {test_accuracy:.4f}")

 # Bouns
# Prediction and Manual Accuracy
print("\nManual Accuracy Check on 10 Random Samples ")
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'] # list contain names

indices = random.sample(range(len(x_test)), 10) # random from 0 to 10 (from calss_name)
predictions = model.predict(x_test[indices])
# enumerate like dectionary take {key,value} i = key and idx = value (bag,shirt,gress,etc..)
for i, idx in enumerate(indices): # to loop in indices and get the manual accuracy
    pred_label = np.argmax(predictions[i])
    true_label = y_test[idx]
    check = " Correct" if pred_label == true_label else " Incorrect" # ternary operator like
    print(f"Sample {i+1}: Predicted: {class_names[pred_label]}, True: {class_names[true_label]} → {check}")

# BONUS
# Retrain with SGD optimizer
model_sgd = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(784,)),
tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax') # Softmax is used to find the probabilities.
])

model_sgd.compile(optimizer='sgd',
                  loss='sparse_categorical_crossentropy', # دالداله  lose function
                  metrics=['accuracy'])

model_sgd.fit(x_train, y_train, epochs=10, validation_split=0.1) #  validation_splite (  عدد المرات الي هتحقق منها ف الموديل 10% 1)
test_loss_sgd, test_accuracy_sgd = model_sgd.evaluate(x_test, y_test)
print(f"\n sgd Model Accuracy on Test Data: {test_accuracy_sgd:.4f}")
print(f"Difference in Accuracy: {test_accuracy - test_accuracy_sgd:.4f}")
```
